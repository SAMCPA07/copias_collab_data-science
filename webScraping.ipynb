```python
import requests
from bs4 import BeautifulSoup
import pandas as pd  # Importamos pandas

url = 'https://es.wikipedia.org/wiki/Anexo:Episodios_de_Gravity_Falls'

# El "disfraz" (User-Agent) para que Wikipedia no nos bloquee
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'
}

respuesta = requests.get(url, headers=headers)
soup = BeautifulSoup(respuesta.text, 'html.parser')

# Buscamos las tablas de Wikipedia
tablas = soup.find_all('table', class_='wikitable')

# Creamos una lista vacía donde iremos guardando los datos
datos_episodios = []

for tabla in tablas:
    # Obtenemos todas las filas, saltando la primera (los encabezados)
    filas = tabla.find_all('tr')[1:]

    for fila in filas:
        celdas = fila.find_all(['th', 'td'])

        # Filtramos para asegurarnos de que la fila tenga suficientes columnas
        if len(celdas) >= 3:
            numero = celdas[0].text.strip()

            # Verificamos que sea un número de episodio válido
            if numero.isdigit() or numero.replace('a', '').replace('b', '').isdigit():

                # Limpiamos el título de referencias como [1]
                titulo = celdas[2].text.strip().split('[')[0].strip()

                # ¡Aquí está el truco! Agregamos los datos como un diccionario a nuestra lista
                datos_episodios.append({'Número': numero, 'Título': titulo})

# --- LA MAGIA DE PANDAS ---
# Convertimos nuestra lista de diccionarios en un DataFrame
df = pd.DataFrame(datos_episodios)

print("--- DATAFRAME DE EPISODIOS DE GRAVITY FALLS ---")
# Imprimimos el DataFrame (pandas automáticamente le dará formato de tabla)
print(df.to_string(index=False)) # Usamos to_string para que te imprima todas las filas y ocultamos el índice
```

Este es la version con diccionarios, en este ya defini los nombres de los alumjos conjunto a su calificacion


```python
def promedio_general(alumnoscali):
    suma = sum(alumnoscali.values())
    promedio = suma / len(alumnoscali)

    print(f"El promedio general de las calificaciones es de {promedio:.2f}")
    return promedio

def cali_mas_alta_baja(alumnoscali):
    maximo = max(alumnoscali.values())
    minimo = min(alumnoscali.values())
    print(f"La calificacion mas alta es de {maximo:.2f} y la mas baja es de {minimo:.2f}")

def aprobados_reprobados(alumnoscali):
    aprobados = []
    reprobados = []

    for nombre, calificacion in alumnoscali.items():
        if calificacion >= 60:
            aprobados.append(nombre.upper())
        else:
            reprobados.append(nombre.upper())

    aprobados_formateados = ", ".join(aprobados)
    reprobados_formateados = ", ".join(reprobados)

    print(f"Los alumnos aprobados son: {aprobados_formateados}")
    print(f"Los alumnos reprobados son: {reprobados_formateados}")

def estadistica(alumnoscali):
    for nombre, cal in alumnoscali.items():
        if 90 <= cal <= 100:
            print(f"La calificacion de {nombre} ({cal}) es Excelente")
        elif 80 <= cal <= 89:
            print(f"La calificacion de {nombre} ({cal}) es Bueno")
        elif 70 <= cal <= 79:
            print(f"La calificacion de {nombre} ({cal}) es Suficiente")
        elif cal < 70:
            print(f"La calificacion de {nombre} ({cal}) es Reprobado")


print("--- Calificaciones de Alumnos ---")

diccionario_alumnos = {
    "Luis": 95.0,
    "Jesus": 85.5,
    "Gael": 75.0,
    "Giovanni": 55.0,
    "Emiliano": 92.0,
    "Angelo": 65.0,
    "Anuel": 88.0,
    "Oscar": 100.0
}

aprobados_reprobados(diccionario_alumnos)
print("-" * 40)
cali_mas_alta_baja(diccionario_alumnos)
print("-" * 40)
promedio_general(diccionario_alumnos)
print("-" * 40)
estadistica(diccionario_alumnos)
```

Aqui va el ejercicio de la floreria, donde se tiene que checar porque no imprimio la cantidad debida

```python
flores = {
    "rosas": 10,
    "tulipanes": 100,
    "girasol": 5,
    "margaritas":90
}

# imprimir claves
for flor in flores.keys():
    print(flor)
# imprimir valores
for cantidad in flores.values():
    print(cantidad)

# imprmir clave-valor
for flor, cantidad in flores.items():
    if cantidad < 10:
        print(f"comprar mas {flor}: {cantidad}: ")
        comprar = input(f"deseas comprar mas {flor} s = si | n = no?: ")

        if comprar != "s":
            break
        else:
            compra = int(input(f"cuantas {flor} deseas comprar?: "))

            flores[flor] += compra

            print(f"stock {flor}: {flores[flor]}")

    else:
        print(f"suficiente stock de {flor}: {cantidad}")

print(flores)
```

Lo arreglé quitando el .values() y el .update(). En vez de eso, accedí directo al diccionario usando corchetes: puse flores[flor] += compra. Así le sumo la cantidad comprada directamente al stock de esa flor y el código queda mucho más limpio.
